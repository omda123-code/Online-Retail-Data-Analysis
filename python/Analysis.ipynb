```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import os
```

```python
os.makedirs('visuals',exist_ok=True)
```

```python
# 1. Total Sales Revenue
total_sales = pd.read_csv("C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Online Retail/outputs/total_sales.csv")

plt.figure(figsize=(4,4))
plt.bar(['Total Revenue'], total_sales['TotalRevenue'])
plt.title("Total Sales Revenue")
plt.ylabel("Revenue (£)")
plt.savefig("visuals/total_sales.png")
plt.show()
```

```python
# 2. Monthly Sales Trend
monthly_sales = pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Online Retail/outputs/monthly_sales_trend.csv')
plt.figure(figsize=(12,6))
sns.lineplot(
    data=monthly_sales, 
    x='SalesMonth', y='MonthlyRevenue', hue='SalesYear', marker='o'
)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue (£)")
plt.legend(title='Year')
plt.savefig("visuals/monthly_sales_trend.png")
plt.show()
```

```python
# 3. Top 10 Countries by Revenue
top_countries = pd.read_csv("C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Online Retail/outputs/top_countries.csv")
plt.figure(figsize=(10,6))
sns.barplot(data=top_countries, x='Revenue', y='Country',hue='Country', palette="viridis")
plt.title("Top 10 Countries by Revenue")
plt.xlabel("Revenue (£)")
plt.ylabel("Country")
plt.savefig("visuals/top_countries.png")
plt.show()
```

```python
# 4. Top 10 Customers by Revenue
top_customers = pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Online Retail/outputs/top_customers.csv')
plt.figure(figsize=(10,6))
sns.barplot(data=top_customers, x='Revenue', y='CustomerID',hue='CustomerID', palette="magma")
plt.title("Top 10 Customers by Revenue")
plt.xlabel("Revenue (£)")
plt.ylabel("Customer ID")
plt.xticks(rotation=90)
plt.savefig("visuals/top_customers.png")
plt.show()
```

```python
# 5. Best-Selling Products
best_products = pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Online Retail/outputs/best_selling_products.csv')
plt.figure(figsize=(12,8))
sns.barplot(data=best_products, x='TotalQuantity', y='Description', hue='Description',palette="cubehelix")
plt.title("Best-Selling Products by Quantity")
plt.xlabel("Total Quantity")
plt.ylabel("Product Description")
plt.savefig("visuals/best_selling_products.png")
plt.show()
```

```python
# 6. Top Revenue Products
top_revenue_products = pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Online Retail/outputs/top_revenue_products.csv')
plt.figure(figsize=(12,8))
sns.barplot(data=top_revenue_products, x='Revenue', y='Description', palette="coolwarm")
plt.title("Top Revenue Products")
plt.xlabel("Revenue (£)")
plt.ylabel("Product Description")
plt.savefig("visuals/top_revenue_products.png")
plt.show()
```

```python
# 8. Cancelled Orders
cancelled_orders = pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Online Retail/outputs/cancelled_orders.csv')
plt.figure(figsize=(4,4))
plt.bar(['Cancelled Orders'], cancelled_orders['CancelledOrders'])
plt.title("Number of Cancelled Orders")
plt.ylabel("Orders Count")
plt.savefig("visuals/cancelled_orders.png")
plt.show()
```
